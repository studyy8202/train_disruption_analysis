```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/disruptions-2023.csv')
```

```python
# Display basic information
print(df.info())
print(df.describe())
```

```python
# Convert start_time and end_time to datetime
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])
```

```python
df['hour'] = df['start_time'].dt.hour
df['day_of_week'] = df['start_time'].dt.dayofweek
```

```python
# Most common disruption causes
common_causes = df['cause_en'].value_counts().head(10)
print("Most Common Disruption Causes:\n", common_causes)
```

```python
# Average duration by disruption cause
avg_duration_cause = df.groupby('cause_en')['duration_minutes'].mean().sort_values(ascending=False).head(10)
print("Average Duration by Cause:\n", avg_duration_cause)
```

```python
# Total duration by disruption category
total_duration_category = df.groupby('cause_group')['duration_minutes'].sum().sort_values(ascending=False)
print("Total Duration by Category:\n", total_duration_category)
```

```python
# Disruption distribution by hour
disruption_by_hour = df['hour'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.barplot(x=disruption_by_hour.index, y=disruption_by_hour.values)
plt.title('Disruptions by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Disruptions')
plt.show()
```

```python
# Disruption distribution by day of the week
disruption_by_day = df['day_of_week'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.barplot(x=disruption_by_day.index, y=disruption_by_day.values)
plt.title('Disruptions by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Disruptions')
plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```

```python
#Routes most affected by disruptions

disruptions_by_route = df['ns_lines'].value_counts().head(10)
print("Top 10 Routes Most Affected by Disruptions:\n", disruptions_by_route)
```

```python
# Find the most common disruption cause for each of the top 10 routes
for route in disruptions_by_route.index:
    route_df = df[df['ns_lines'] == route]
    common_cause = route_df['cause_en'].value_counts().idxmax()
    print(f"Most common disruption cause for route {route} is: {common_cause}")

# Plotting the top 10 routes and their disruption counts
plt.figure(figsize=(12, 6))
sns.barplot(x=disruptions_by_route.values, y=disruptions_by_route.index)
plt.title('Top 10 Routes with the Most Disruptions')
plt.xlabel('Number of Disruptions')
plt.ylabel('Route')
plt.show()
```

```python
df['month'] = df['start_time'].dt.to_period('M')
monthly_disruptions = df.groupby('month').size()
plt.figure(figsize=(14, 7))
monthly_disruptions.plot()
plt.title('Monthly Disruptions Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Disruptions')
plt.show()
```

```python
df['severity'] = pd.cut(df['duration_minutes'], bins=[0, 30, 60, 120, float('inf')], labels=['Low', 'Medium', 'High', 'Critical'])
severity_counts = df['severity'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=severity_counts.index, y=severity_counts.values)
plt.title('Disruptions by Severity')
plt.xlabel('Severity')
plt.ylabel('Number of Disruptions')
plt.show()
```

```python
route_impact = df.groupby('ns_lines')['duration_minutes'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(14, 7))
sns.barplot(x=route_impact.values, y=route_impact.index)
plt.title('Top 10 Routes by Total Disruption Duration')
plt.xlabel('Total Disruption Duration (minutes)')
plt.ylabel('Route')
plt.show()

station_impact = df['rdt_station_names'].str.split(',', expand=True).stack().value_counts().head(10)
plt.figure(figsize=(14, 7))
sns.barplot(x=station_impact.values, y=station_impact.index)
plt.title('Top 10 Stations by Number of Disruptions')
plt.xlabel('Number of Disruptions')
plt.ylabel('Station')
plt.show()
```

```python
# Severity by cause
severity_by_cause = df.groupby(['cause_en', 'severity']).size().unstack().fillna(0)
print("Severity by Cause:\n", severity_by_cause)

# Plot severity by cause
severity_by_cause.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title('Severity by Disruption Cause')
plt.xlabel('Cause')
plt.ylabel('Number of Disruptions')
plt.legend(title='Severity')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df[['duration_minutes', 'hour', 'day_of_week']].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Preparing the dataset for predictive modeling
features = df[['hour', 'day_of_week', 'ns_lines']]
target = df['cause_group']

# Encoding categorical features
features = pd.get_dummies(features, columns=['ns_lines'])

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Training the model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions and evaluation
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
